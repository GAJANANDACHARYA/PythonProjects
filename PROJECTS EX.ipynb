# To-Do List

```python
# Define an empty list to store tasks
tasks = []

# Function to add a task
def add_task(task):
    tasks.append(task)
    print(f"Task '{task}' added successfully!")

# Function to remove a task
def remove_task(task):
    if task in tasks:
        tasks.remove(task)
        print(f"Task '{task}' removed successfully!")
    else:
        print(f"Task '{task}' not found!")

# Function to mark a task as completed
def complete_task(task):
    if task in tasks:
        tasks.remove(task)
        completed_task = f"[Completed] {task}"
        tasks.append(completed_task)
        print(f"Task '{task}' marked as completed!")
    else:
        print(f"Task '{task}' not found!")

# Function to display the to-do list
def show_tasks():
    if tasks:
        print("\nTo-Do List:")
        for index, task in enumerate(tasks, start=1):
            print(f"{index}. {task}")
    else:
        print("\nYour to-do list is empty!")

# Main loop
while True:
    print("\nOptions:")
    print("1. Add Task")
    print("2. Remove Task")
    print("3. Mark Task as Completed")
    print("4. Show Tasks")
    print("5. Quit")

    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == '1':
        task = input("Enter the task to add: ")
        add_task(task)
    elif choice == '2':
        task = input("Enter the task to remove: ")
        remove_task(task)
    elif choice == '3':
        task = input("Enter the task to mark as completed: ")
        complete_task(task)
    elif choice == '4':
        show_tasks()
    elif choice == '5':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please select a valid option (1/2/3/4/5).")
```

# Calculator

```python
# Function to perform addition
def add(x, y):
    return x + y

# Function to perform subtraction
def subtract(x, y):
    return x - y

# Function to perform multiplication
def multiply(x, y):
    return x * y

# Function to perform division
def divide(x, y):
    if y == 0:
        return "Cannot divide by zero!"
    return x / y

# Main calculator loop
while True:
    print("\nOptions:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Quit")

    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == '5':
        print("Goodbye!")
        break

    if choice not in ('1', '2', '3', '4'):
        print("Invalid choice. Please select a valid option (1/2/3/4).")
        continue

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == '1':
        print(f"Result: {num1} + {num2} = {add(num1, num2)}")
    elif choice == '2':
        print(f"Result: {num1} - {num2} = {subtract(num1, num2)}")
    elif choice == '3':
        print(f"Result: {num1} * {num2} = {multiply(num1, num2)}")
    elif choice == '4':
        result = divide(num1, num2)
        if isinstance(result, str):
            print(result)
        else:
            print(f"Result: {num1} / {num2} = {result}")
```

# Q&A Program

```python
list1 = [("Capital of India?", "Delhi"),("Ronaldo Jersey Number?", "7"),("Rohit Sharma Jersey Number?", "45"),
        ("What is capital of france","Paris"),("What is the largest planet in our solar system?","Jupiter"),("Who wrote the play 'Romeo and Juliet'?","William Shakespeare")]
score=0
def ask_questions():
    global score
    for question, answer in list1:
        user_answer = input(question + "\nYour answer: ")
        if user_answer.lower() == answer.lower():
            print("Correct!\n")
            score+=100
        else:
            print("Incorrect. The correct answer is:", answer, "\n")
def main():
    print("Welcome to the Python Q&A Quiz!\n")
    ask_questions()
    print("Your Score is", score)
    print("Thanks for playing!")

if __name__ == "__main__":
    main()
```

# Q&A 2nd Program

```python
questions = ["What is capital of India?","Rohit Sharma jersey number?","Ronaldo jersey number?","What is the capital of France?","What is the largest planet in our solar system?","Who wrote the play 'Romeo and Juliet'?",]

answers = ["Delhi","45","7","Paris","Jupiter","William Shakespeare",]
score = 0
print("Welcome to the Python Q&A Quiz!\n")
def ask_question(question, correct_answer):
    global score
    user_answer = input(f"{question}\n")

    if user_answer.lower() == correct_answer.lower():
        print("Correct!")
        score += 100
    else:
        print("Incorrect. The correct answer is:", correct_answer)

for i in range(len(questions)):
    ask_question(questions[i], answers[i])
print("Your final score is:", score)
print("Thanks For Playing")
```
